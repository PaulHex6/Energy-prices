```python
import requests
import json
from datetime import datetime

def get_demand_data(business_date):
    # Definiowanie URL i parametrów zapytania
    url = "https://api.raporty.pse.pl/api/his-obc"
    params = {
        "$filter": f"business_date eq '{business_date}'"
    }

    # Wykonanie zapytania GET do API
    response = requests.get(url, params=params)

    # Sprawdzenie statusu odpowiedzi
    if response.status_code == 200:
        # Parsowanie odpowiedzi jako JSON
        data = response.json()
        return data['value']
    else:
        print(f"Błąd: {response.status_code}")
        return None

# Przykład użycia funkcji
business_date = "2024-07-01"
demand_data = get_demand_data(business_date)

# Wyświetlenie danych
if demand_data:
    for entry in demand_data:
        print(f"Czas: {entry['udtczas']}, Zapotrzebowanie: {entry['obciazenie']} MW")

```
